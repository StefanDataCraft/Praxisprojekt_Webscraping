```python
# Die üblichen verdächtigen Imports
import requests
import os
import pandas as pd
from bs4 import BeautifulSoup

url_anfang = 'https://books.toscrape.com/'

# Komplette Schleife für zehn nachfolgende Seiten (deluxe Version)
full_list = []
for p in range(1, 11):
    full_url = url_anfang + f'catalogue/page-{p}.html'
    response = requests.get(full_url)
    soup = BeautifulSoup(response.content, 'html.parser')

    for book_box in soup.find_all("article", class_="product_pod"):
        title = book_box.find('h3').find('a')['title']
        rating = book_box.find('p')['class'][1]
        price = book_box.find('div', class_="product_price").find('p', "price_color").text
        clean_price = price.replace("£","")
        full_list.append({"Title": title, "Rating": rating, "Price": clean_price})


print(full_list)
df = pd.DataFrame(full_list, columns=["Title", "Rating", "Price"])
print(os.getcwd())
df.to_csv("./book_data.csv", index=False)
```
